## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
year = 2020
if year % 4 != 0: print("Обычный год")
elif year % 100 == 0: 
    if year % 400 == 0: print("Високоснй год")
    else: print("Обычный год")
else: print("Високосный год")
````

## Задание 4

Вам нужно написать программу для подбора упаковок по размерам товара. Размеры хранятся в переменных (в сантиметрах):

```
width = 10
length = 205
height = 5
```

Используйте следующие правила:
- если каждое из трех измерений менее 15 сантиметров, то выведите на экран "Коробка №1";
- если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
- если длина товара больше 2 метров, то выводите "Упаковка для лыж";
- во всех остальных случаях выводите "Стандартная коробка №3".


````python
width = 10
length = 205
height = 5

if width < 15 and length < 15 and height < 15: print("Коробка №1")
elif width > 15 and width < 50 or length > 15 and length < 50 or height > 15 and height < 50: print("Коробка №2")
elif length > 200: print("Упаковка для лыж")
else: print("Стандартная коробка №3")
````

## Задание 5

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
number = 123321

numberStr = str(number)

sumFirst = 0
sumSecond = 0

for i in range(6):
    if i < 3:
        sumFirst += int( numberStr[i] )
    else:
        sumSecond += int( numberStr[i] )
        
if sumFirst == sumSecond: print("Счастливый билет")
else: print("Несчастливый билет")
````

## Задание 6

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
import math

def circle():
    print("Введите радиус круга:")
    r = int( input() )
    square = math.pi * r * r
    print( "Площадь круга: " + str(square) )
    
def triangle():
    print("Введите длину стороны A:")
    a = int( input() )
    print("Введите длину стороны B:")
    b = int( input() )
    print("Введите длину стороны C:")
    c = int( input() )
    halfPerimetr = (a + b + c) / 2;
    square = math.sqrt( halfPerimetr * (halfPerimetr - a) * (halfPerimetr - b) * (halfPerimetr - c) )
    print( "Площадь треугольника: " + str(square) )
    
def rectangle():
    print("Введите длину стороны A:")
    a = int( input() )
    print("Введите длину стороны B:")
    b = int( input() )
    square = a * b
    print( "Площадь прямоугольника: " + str(square) )

print("Введите тип фигуры:")
type = input()

if type == "Круг":
    circle()
elif type == "Треугольник":
    triangle()
elif type == "Прямоугольник":
    rectangle()
````
